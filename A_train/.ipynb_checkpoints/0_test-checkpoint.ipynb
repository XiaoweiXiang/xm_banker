```python
import os
import sys
sys.path.append('..')
import numpy as np
import pandas as pd
from library import config, load_o_data
```

```python
df_test = load_o_data.load_df_test()
```

```python
df_trade = load_o_data.load_df_trade()
```

```python
df_test_trade = pd.merge(df_test[['core_cust_id', 'prod_code', 'deal_date']], df_trade, on=['core_cust_id', 'prod_code'], how='left', suffixes=('', '_y'))
df_test_trade = df_test_trade.loc[df_test_trade['deal_date'] > df_test_trade['deal_date_y'], ]
df_test_trade_g = df_test_trade.groupby(['core_cust_id', 'prod_code', 'deal_date']).agg({'trade_id': len}).reset_index()
df_test_trade_g
```

```python
df_test_predict = pd.merge(df_test, df_test_trade_g, on=['core_cust_id', 'prod_code', 'deal_date'], how='left')
df_test_predict['y'] = [1-int(x) for x in pd.isnull(df_test_predict['trade_id'])]
df_test_predict
```

```python
df_test_predict[['id', 'core_cust_id', 'prod_code', 'y']].to_csv(os.path.join(config.result_p_dir, 'his_20210109.csv'), index=False)
```
